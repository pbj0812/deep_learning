```python
column_names = result_table.columns
print(column_names)

column_length = len(column_names)
print(column_length)
```

```python
normal_table = result_table.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler(feature_range = (0, 1))
```

```python
# batch normalization
for i in range(1,column_length):
    test_data = np.array(normal_table[result_table.columns[i]])
    test_data = np.reshape(test_data,[-1,1])
    a = sc.fit_transform(test_data)
    normal_table[column_names[i]] = a
```

```python
dataset = normal_table[['date', 'Solar_radiation(MJ/m^2)', 'Sunshine(Sec)', 'temp', 'hum(%)', 'InteriorEquipment:Electricity [J](Hourly)', 'pw_light']]
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
Train_dataset = dataset.loc[:6480]
Test_dataset = dataset.loc[6480:]
```

```python
Train_dataset.head()
```

```python
Test_dataset.head()
```

```python
# define train dataset
```

```python
train_x = Train_dataset.iloc[:,1:-1]
```

```python
train_x.head()
```

```python
train_y = Train_dataset.iloc[:,-1]
```

```python
train_y.head()
```

```python
train_x = np.array(train_x)
train_x = np.reshape(train_x, (train_x.shape[0], train_x.shape[1], 1))
```

```python
train_y = np.array(train_y)
train_y = np.reshape(train_y, (train_y.shape[0], 1))
```

```python
# define test dataset
```

```python
test_x = Test_dataset.iloc[:, 1:-1]
```

```python
test_x.head()
```

```python
test_y = Test_dataset.iloc[:, -1]
```

```python
test_y.head()
```

```python
test_x = np.array(test_x)
test_x = np.reshape(test_x, (test_x.shape[0], test_x.shape[1], 1))
```

```python
test_y = np.array(test_y)
test_y = np.reshape(test_y, (test_y.shape[0], 1))
```

```python
# define model
```

```python
regressor = Sequential()
regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (train_x.shape[1], 1)))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))
regressor.add(Dense(units = 1))
```

```python
regressor.compile(optimizer = 'adam', loss = 'mae') # mean_squared_error
```

```python
regressor.fit(train_x, train_y, epochs = 100, batch_size = 32)
```

```python
predict = regressor.predict(test_x)
```

```python
import matplotlib.patches as mpatches
```

```python
plt.plot(predict, color = 'red', label = 'Predict')
plt.plot(test_y, color = 'blue', label = 'Real')

red_patch = mpatches.Patch(color='red', label='LSTM')
blue_patch = mpatches.Patch(color='blue', label='REAL')
plt.legend(handles=[blue_patch, red_patch])
```
