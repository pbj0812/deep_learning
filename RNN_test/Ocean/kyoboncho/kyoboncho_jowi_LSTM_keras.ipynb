```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read data file
```

```python
data = pd.read_table('./dataset/kyoboncho_2017.txt', na_values='-') 
```

```python
data
```

```python
plt.plot(data.jowi)
```

```python
# define dataset
```

```python
tide_level = np.array(data.jowi) # df2array
print(np.shape(tide_level))
```

```python
len(tide_level) # length of data
```

```python
dataset = tide_level[:10000]
```

```python
# check nan_value index
import math

nan_index = []

for i in range(len(dataset)):
    if(math.isnan(dataset[i]) == True):
        nan_index.append(i)
print(len(nan_index))
print(nan_index)
```

```python
# interp nan value
for i in nan_index:
    cnt = 1
    while (math.isnan(dataset[i+cnt]) == True):
        cnt += 1
    if (dataset[i-1] < dataset[i+cnt+1]):
        dataset[i] = np.linspace(dataset[i-1], dataset[i+cnt+1], cnt+3)[1]
    elif (dataset[i-1] > dataset[i+cnt+1]):
        dataset[i] = np.linspace(dataset[i+cnt+1], dataset[i-1], cnt+3)[1]
    else:
        dataset[i] = dataset[i-1]
```

```python
# re check nan_value index
import math

nan_index = []

for i in range(len(dataset)):
    if(math.isnan(dataset[i]) == True):
        nan_index.append(i)
if len(nan_index)==0:
    print('OK')
```

```python
train_dataset = dataset[:9000]
test_dataset = dataset[9000:]
```

```python
plt.plot(train_dataset)
```

```python
plt.plot(test_dataset)
```

```python
# scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
train_dataset = np.reshape(train_dataset, (np.shape(train_dataset)[0], 1))
test_dataset = np.reshape(test_dataset, (np.shape(test_dataset)[0], 1))
print(np.shape(train_dataset))
print(np.shape(test_dataset))
```

```python
sc = MinMaxScaler(feature_range = (0, 1))
train_dataset = sc.fit_transform(train_dataset)
test_dataset = sc.fit_transform(test_dataset)
```

```python
# define train_dataset
```

```python
train_X = []
train_y = []
```

```python
for i in range(50, len(train_dataset)):
    train_X.append(train_dataset[i-50:i])
    train_y.append(train_dataset[i])
```

```python
print(np.shape(train_X), np.shape(train_y))
```

```python
train_X = np.array(train_X)
train_X = np.reshape(train_X, (train_X.shape[0], train_X.shape[1], 1))
train_y = np.array(train_y)
```

```python
print(np.shape(train_X))
print(np.shape(train_y))
```

```python
# define test_datset
```

```python
test_X = []
test_y = []
```

```python
for i in range(50, len(test_dataset)):
    test_X.append(test_dataset[i-50:i])
    test_y.append(test_dataset[i])
```

```python
print(np.shape(test_X), np.shape(test_y))
```

```python
test_X = np.array(test_X)
test_X = np.reshape(test_X, (test_X.shape[0], test_X.shape[1], 1))
test_y = np.array(test_y)
```

```python
print(np.shape(test_X))
print(np.shape(test_y))
```

```python
# Building the LSTM
```

```python
# import package
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
# initializing the LSTM
```

```python
regressor = Sequential()
```

```python
# adding the first to the fifth LSTM layer and some dropout regularization
```

```python
# first
regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (train_X.shape[1], 1)))
regressor.add(Dropout(0.2))
# second
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
# third
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
# fourth
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
# fifth
regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))
```

```python
# adding the output layer
```

```python
regressor.add(Dense(units = 1))
```

```python
# compiling the RNN
```

```python
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# fitting the RNN to the Training set
```

```python
regressor.fit(train_X, train_y, epochs = 100, batch_size = 32)
```

```python
predict = regressor.predict(test_X)
```

```python
import matplotlib.patches as mpatches
plt.plot(predict, color = 'red', label = 'Predict')
plt.plot(test_y, color = 'blue', label = 'Real')
plt.ylabel('tide_level(normalization)')
red_patch = mpatches.Patch(color='red', label='Predict')
blue_patch = mpatches.Patch(color='blue', label='Real')
plt.legend(handles=[red_patch, blue_patch])
```
