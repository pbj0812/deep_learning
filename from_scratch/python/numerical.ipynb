```python
def numerical_diff(f, x):
    h = 1e-4
    return (f(x+h) - f(x-h)) / (2*h)
```

```python
def function_1(x):
    return 0.01*x**2 + 0.1*x
```

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y)
plt.show()
```

```python
def numerical_gradient(f, x):
    h = 1e-4
    grad = np.zeros_like(x)
    
    print(grad)
    
    for idx in range(x.size):
        print(x.size)
        
        tmp_val = x[idx]
        print(tmp_val)
        
        x[idx] = tmp_val + h
        print(x[idx])
        
        fxh1 = f(x)
        print(fxh1)
        
        x[idx] = tmp_val - h
        print(x[idx])
        
        fxh2 = f(x)
        print(fxh2)
        
        grad[idx] = (fxh1 - fxh2) / (2*h)
        print(grad[idx])
        
        x[idx] = tmp_val
        print(x[idx])
        print(x)
        
        print(grad)
        
        
        print('####################')
        
    return grad
```

```python
def gradient_descent(f, init_x, lr =0.01, step_num=100):
    x = init_x
    
    for i in range(step_num):
        grad = numerical_gradient(f, x)
        x -= lr*grad
        print('error : ', x)
    return x
```
